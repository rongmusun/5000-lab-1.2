## Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("penguins")
print(df)

# INSERT YOUR SOLUTION HERE
sns.set_theme(style='white',palette='Pastel1')
FS=18
g=sns.relplot(
    data=df,
    x='bill_length_mm',
    y='flipper_length_mm',
    hue='species',
    col='island',
    s=200
)
print(type(g))
g.axes[0,0].set_xlabel('Penguin bill length (mm)', fontsize=FS)
g.axes[0,1].set_xlabel('Penguin bill length (mm)', fontsize=FS)
g.axes[0,2].set_xlabel('Penguin bill length (mm)', fontsize=FS)
g.axes[0,0].set_ylabel('Penguin flipper length (mm)', fontsize=FS)
g.axes[0,0].set_title("Torgersen Island",fontsize=FS)
g.axes[0,1].set_title("Biscoe Island",fontsize=FS)
g.axes[0,2].set_title("Dream Island",fontsize=FS)
for i in range(0,3):
    g.axes[0,i].tick_params(axis='both', which='major', labelsize=FS)

plt.setp(g._legend.get_texts(), fontsize=FS)
plt.setp(g._legend.get_title(), fontsize=FS)

plt.savefig('plot-1.pdf', bbox_inches='tight')
```
